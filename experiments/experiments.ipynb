```python
import torch
from torch.optim import AdamW
from transformers import GPT2Config, GPT2Tokenizer, GPT2LMHeadModel, get_linear_schedule_with_warmup
```

```python
torch.cuda.is_available()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
text = "Who are you?."
```

```python
ids = torch.tensor(tokenizer.encode([text]))
ids
```

```python
output_ids = model.generate(ids)
output_ids
```

```python
tokenizer.decode(output_ids)
```

```python
from transformers import AutoModelForCausalLM
```

```python
model = AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
ids
```

```python
ids
```

```python
out_ids = model.forward(ids).logits
out_ids.shape
```

```python
last_out_id = out_ids[0,-1,:]
last_out_id.shape
```

```python
out_prob = torch.softmax(last_out_id,dim=-1)
out_prob
```

```python
out_id = torch.argmax(out_prob)
out_id
```

```python
tokenizer.decode(out_id)
```
